```python
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
```

```python
# df = https://www.openintro.org/data/csv/loans_full_schema.csv
df = pd.read_csv("data/loans_full_schema.csv")
df.head(10)
```

```python
df = df.drop(columns = ["emp_title"])
df.head(5)
```

```python
h2o.init()
```

```python
df_aml = h2o.H2OFrame(df)
```

```python
df_aml.describe()
```

```python
y = "loan_amount"
```

```python
splits = df_aml.split_frame(ratios = [0.8], seed = 666)
train = splits[0]
test = splits[1]
```

```python
aml = H2OAutoML(max_runtime_secs = 1200, seed = 666, project_name = "autoML_rep_w_py")
aml.train(y = y, training_frame = train, leaderboard_frame = test)
```

```python
aml.leaderboard.head()
```

```python
# pred = aml.predict(test)
```

```python
perf_metr = aml.leader.model_performance(test)
perf_metr
```
